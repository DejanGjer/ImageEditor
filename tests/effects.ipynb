```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

print("Setup completed!")
```

```python
image = plt.imread('elementi.bmp').astype(np.uint8)
print(image.shape)
```

```python
plt.imshow(image)
plt.show()
```

```python
# Convert image to HSV color space for saturation adjustment
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
h, s, v = cv2.split(hsv_image)
value = 200

lim = 255 - value
v[v > lim] = 255
v[v <= lim] += value

final_hsv = cv2.merge((h, s, v))
final_hsv
```

```python
adjusted_image = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR).astype(np.uint8)

# Plot original and adjusted image
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(adjusted_image)
plt.title('Adjusted Image')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def adjust_brightness(image, brightness):
    # Convert brightness to a factor between -100 and 100
    brightness_factor = 1.0 + (brightness / 100.0)

    # Increase the brightness
    brightened_image = np.clip(image * brightness_factor, 0, 255).astype(np.uint8)

    return brightened_image

# Load the RGB image
image = cv2.imread('elementi.bmp')

# Specify the brightness value between -100 and 100
brightness_value = 20  # You can adjust this value as needed

# Adjust the brightness and get the result
brightened_image = adjust_brightness(image, brightness_value)

brightened_image = cv2.cvtColor(brightened_image, cv2.COLOR_BGR2RGB)

# Display the images
# cv2.imshow('Original image', image)
# cv2.imshow('Brightened image', brightened_image)
# cv2.waitKey(0)
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(brightened_image)
plt.title('Adjusted Image')
plt.axis('off')
plt.show()


# Save the new image
cv2.imwrite('brightened_image.png', brightened_image)
```
